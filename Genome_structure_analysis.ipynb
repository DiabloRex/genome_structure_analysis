```python
%%cython
from libc.math cimport abs, sqrt, ceil, floor, log, log2, acos, cos
from numpy cimport ndarray
import numpy as np
from numpy import ones, zeros, int32, float32, uint8, fromstring
from numpy import sort, empty, array, arange, concatenate, searchsorted

def pairRegionsIntersection(ndarray[int, ndim=2] pairs,
                            ndarray[int, ndim=2] regions,
                            exclude=False, allow_partial=False,
                            region_indices=False):
  
  cdef int i, j, k, a, b
  cdef int exc = int(exclude)
  cdef int partial = int(allow_partial)
  cdef int ni = 0
  cdef int np = len(pairs)
  cdef int nr = len(regions)
  cdef ndarray[int, ndim=1] indices = empty(np, int32)
  cdef ndarray[int, ndim=1] indices_reg = empty(np, int32)
  cdef ndarray[int, ndim=1] order = array(regions[:,0].argsort(), int32)  
  
  for i in range(np):
    
    if pairs[i,1] < regions[order[0],0]:
      if exc:
        indices[ni] = i
        ni += 1
      
      continue

    if pairs[i,0] < regions[order[0],0]:
      if exc and partial:
        indices[ni] = i
        ni += 1
      
      continue
      
    a = 0
    b = 0
    
    for k in range(nr):
      j = order[k]
      #print i, j, k
      
      if (regions[j,0] <= pairs[i,0]) and (pairs[i,0] <= regions[j,1]):
        a = 1
    
      if (regions[j,0] <= pairs[i,1]) and (pairs[i,1] <= regions[j,1]):
        b = 1
 
      if (pairs[i, 0] < regions[j, 0]) and (pairs[i, 1] < regions[j, 0]):
        break
      
      if partial & (a | b):
        break
      elif a & b:
        break
    
    if partial:
      if exc and not (a & b):
        indices[ni] = i
        indices_reg[ni] = j
        ni += 1
      
      elif a | b:
        indices[ni] = i
        indices_reg[ni] = j
        ni += 1
      
    else:
      if exc and not (a | b):
        indices[ni] = i
        indices_reg[ni] = j
        ni += 1
      
      elif a & b:
        indices[ni] = i
        indices_reg[ni] = j
        ni += 1

  
  if region_indices:
    return indices[:ni], indices_reg[:ni]
    
    
  else:
    return indices[:ni]

def regionBinValues(ndarray[int, ndim=2] regions, ndarray[double, ndim=1] values,
                    int binSize=1000, int start=0, int end=-1, double dataMax=0.0,
                    double scale=1.0, double threshold=0.0):
                    
  cdef int i, p1, p2, b1, b2, b3, s, e
  cdef int nBins, n = len(values)
  cdef double f, r, v, vMin, vMax
  
  if len(regions) != n:
    data = (len(regions), n)
    raise Exception('Number of regions (%d) does not match number of values (%d)' % data) 
  
  if end < 0:
    end = binSize * int32(regions.max() / binSize)
  
  s = start/binSize
  e = end/binSize
  nBins = 1+e-s
  
  cdef ndarray[double, ndim=1] hist = zeros(nBins, float)
  
  for i in range(n):
    
    v = values[i]
    if abs(v) < threshold:
      continue
    
    if regions[i,0] > regions[i,1]:
      p1 = regions[i,1] 
      p2 = regions[i,0]
    
    else:
      p1 = regions[i,0]
      p2 = regions[i,1]
    
    if end < p1:
      continue
    
    if start > p2:
      continue  
    
    b1 = p1 / binSize
    b2 = p2 / binSize
    
    if b1 == b2:
      if b1 < s:
        continue
      
      if b1 > e:
        continue
        
      hist[b1-s] += v

    else:
      r = <double> (p2-p1)
    
      for b3 in range(b1, b2+1):
        if b3 < s:
          continue
        
        if b3 >= e:
          break  
      
        if b3 * binSize < p1:
          f = <double> ((b3+1)*binSize - p1) / r 
        
        elif (b3+1) * binSize > p2:
          f = <double> (p2 - b3*binSize) / r 
        
        else:
          f = 1.0
      
        hist[b3-s] += v * f
  
  if dataMax != 0.0:
    vMin = hist[0]
    vMax = hist[0]
    
    for i in range(1, nBins):
      if hist[i] < vMin:
        vMin = hist[i]
      
      elif hist[i] > vMax:
        vMax = hist[i]
    
    vMax = max(abs(vMin), vMax, dataMax)

    if vMax > 0.0:
      for i in range(0, nBins):
        hist[i] = hist[i]/vMax
  
  for i in range(0, nBins):
    hist[i] = hist[i] * scale  
  
  return hist
  
def binContacts(ndarray[int, ndim=2] contacts,
                ndarray[int, ndim=2] binMatrix,
                int offsetA, int offsetB, int binSize=2000000,
                int symm=0, int transpose=0):
  
  cdef int i, a, b
  cdef int n, m, nCont = len(contacts[0])
  
  n = len(binMatrix)
  m = len(binMatrix[0])
  
  for i in range(nCont):
    a = (contacts[0,i]-offsetA)/binSize
    b = (contacts[1,i]-offsetB)/binSize
    if transpose:
      a, b = b, a
 
    if (0 <= a < n) and (0 <= b < m):
      binMatrix[a,b] += contacts[2,i]
 
      if symm and (a != b):
        binMatrix[b,a] += contacts[2,i]
  
  return binMatrix

def getInvDistSums(ndarray[double, ndim=2] coords_a,
                   ndarray[double, ndim=2] coords_b,
                   py_seq_pos_a=None,
                   py_seq_pos_b=None,
                   py_min_seq_sep=None, 
                   py_values_b=None,
                   power_adj=1):

  cdef int i, j
  cdef double dx, dy, dz, d
  cdef int p = power_adj
  cdef int na = len(coords_a)
  cdef int nb = len(coords_b)
  cdef long min_seq_sep
  
  cdef ndarray[double, ndim=1] inv_dist_sums = zeros(na, float)
  cdef ndarray[double, ndim=1] values_b
  cdef ndarray[long, ndim=1] seq_pos_a
  cdef ndarray[long, ndim=1] seq_pos_b
  
  if py_values_b is None:
    values_b = ones(nb, float)
  else:
    values_b = py_values_b.astype(float)
  
  if py_min_seq_sep is None:
    for i in range(na):
      for j in range(nb):
        dx = coords_a[i, 0] - coords_b[j, 0]
        dy = coords_a[i, 1] - coords_b[j, 1]
        dz = coords_a[i, 2] - coords_b[j, 2]
        d = dx*dx + dy*dy + dz*dz
        
        if p > 1:
          d = d ** p
        
        if d > 0:
          inv_dist_sums[i] += values_b[j]/d
  
  else:
    min_seq_sep = py_min_seq_sep
    seq_pos_a = py_seq_pos_a
    seq_pos_b = py_seq_pos_b
  
    for i in range(na):
      for j in range(nb):
        if abs(seq_pos_a[i]-seq_pos_b[j]) < min_seq_sep:
          continue
        
        dx = coords_a[i, 0] - coords_b[j, 0]
        dy = coords_a[i, 1] - coords_b[j, 1]
        dz = coords_a[i, 2] - coords_b[j, 2]
        d = dx*dx + dy*dy + dz*dz
        
        if p > 1:
          d = d ** p
 
        if d > 0:
          inv_dist_sums[i] += values_b[j]/d  
  
  return inv_dist_sums
```

### Utility Python functions

```python
from numpy import cov, linalg, sqrt, zeros, random, dot, array, vstack, empty
from random import randint, sample

def kMeansSpread(data, k, thresh=1e-10, verbose=False):
  """k-means with better spread starting values"""
    
  n = len(data)
  index = randint(0, n-1)
  indices = set([index])
  
  influence = zeros(n)
  while len(indices) < k:
    diff = data - data[index]
    sumSq = (diff * diff).sum(axis=1) + 1.0
    influence += 1.0 / sumSq
    index = influence.argmin()
    
    while index in indices:
      index = randint(0, n-1)
    
    indices.add(index)    
  
  centers = vstack([data[i] for i in indices])
    
  return kMeans(data, k, centers, thresh, verbose)


def kMeans(data, k, centers=None, thresh=1e-10, verbose=False):
  """k-means clustering"""
    
  if centers is None:
    centers = array( sample(list(data), k) )  # list() not needed in Python 2

  labels = empty(len(data), float)
  change = 1.0
  prev = []

  j = 0
  while change > thresh:

    clusters = [[] for x in range(k)]
    for i, vector in enumerate(data):
      diffs = centers - vector
      dists = (diffs * diffs).sum(axis=1)
      closest = dists.argmin()
      labels[i] = closest
      clusters[closest].append(vector)
     
    change = 0
    for i, cluster in enumerate(clusters):
      cluster = array(cluster)
      center = cluster.sum(axis=0)/len(cluster)
      diff = center - centers[i]
      change += (diff * diff).sum()
      centers[i] = center
    
    j += 1
    
    if verbose:
      print j, change 
    
  return centers, clusters, labels


def getContactMatrix(contact_dict, chrA, chrB, regionA, regionB, binSize=int(1e6)):
  """Gets a full matrix of contacts from sparse contacts""" 
  
  is_cis = chrA == chrB
    
  startA, endA = regionA
  startB, endB = regionB
  
  s_a = int(startA/binSize)
  s_b = int(startB/binSize)
  e_a = int(endA/binSize)
  e_b = int(endB/binSize)
  
  extentA = endA - startA
  extentB = endB - startB
  
  n = e_a - s_a + 1
  m = e_b - s_b + 1
  
  matrix = zeros((n,m), int32)
  binSize = int32(binSize)
  
  if chrA == chrB:
    chromo_pairs=[(chrA, chrA)]
  else:
    chromo_pairs = [(chrA, chrB), (chrB, chrA)]
    
  for chr1, chr2 in chromo_pairs:
    if chr1 in contact_dict and chr2 in contact_dict[chr1]:
      cData = array(contact_dict[chr1][chr2], int32)
      if chrA == chr1:
        start1, start2 = int32(startA), int32(startB)
        transpose = False
        
      else:
        start1, start2 = int32(startB), int32(startA)
        transpose = True
      
      # Below is additive because both A:B and B:A contacts could be stored
      binContacts(cData, matrix, start1, start2,
                  binSize, is_cis, transpose)
  
  return matrix
    
    
def load_contacts(file_path):
  """Load chromosome and contact data from Numpy archive (.npz)"""
    
  contacts_archive = np.load(file_path)
  chromosomes = set()
  contact_dict = {}
  chromo_limits = {}
    
  for key in contacts_archive:
    null, chr_a, chr_b = key.split('/')
    chromosomes.update([chr_a, chr_b])
    
    if chr_a not in contact_dict:
        contact_dict[chr_a] = {}
    
    contacts  = contacts_archive[key]
    contact_dict[chr_a][chr_b] = contacts
    
    pos_a, pos_b, n_obs = contacts
    
    if chr_a in chromo_limits:
        start, end = chromo_limits[chr_a]
        chromo_limits[chr_a] = (min(start, min(pos_a)), max(end, max(pos_a)))
    else:
        chromo_limits[chr_a] = (min(pos_a), max(pos_a))

    if chr_b in chromo_limits:
        start, end = chromo_limits[chr_b]
        chromo_limits[chr_b] = (min(start, min(pos_b)), max(end, max(pos_b)))
    else:
        chromo_limits[chr_b] = (min(pos_b), max(pos_b))
    
  return chromosomes, chromo_limits, contact_dict


def load_data_track(file_path):
    """Load track data (e.g. ChIp-seq)from Numpy archive (.npz)"""
    
    data_archive = np.load(file_path)
    regions_dict = {}
    values_dict = {}
    
    for key in data_archive:
        null, key2, track_name, chromo = key.split('/')
        
        if key2 == 'regions':
            regions_dict[chromo] = data_archive[key].astype(int32)
        elif key2 == 'values':
            values_dict[chromo] = data_archive[key].astype(float)
    
    return regions_dict, values_dict
    
    
def load_structure_coords(file_path):
    """Load structure coordinates and thier corresponding sequence positions
       from a Numpy (.npz) archive"""
    
    data_archive = np.load(file_path)
    seq_pos_dict = {}
    coords_dict = {}
  
    for key in data_archive:
        key2, chromo = key.split('/')
        
        if chromo.lower()[:3] == 'chr':
            chromo = chromo[3:]
        
        if key2 == 'positions':
            seq_pos_dict[chromo] = data_archive[key].astype(int32)
        elif key2 == 'coords':
            coords_dict[chromo] = data_archive[key].astype(float)
    
    return seq_pos_dict, coords_dict
```

### Make void and A/B compartment data tracks

```python
def calcContactVoidRegions(pop_contacts_path, binSize=int(5e5), close_cis=int(1e6), clip_factor=2.0):

  from numpy import histogram, diag, median
  
  chromosomes, chromo_limits, contact_dict = load_contacts(pop_contacts_path)

  regionDict = {}
  
  for chrA in contact_dict:
    if chrA not in contact_dict[chrA]: # Must have cis data
      continue
    
    limits = chromo_limits[chrA]
    start, end = limits

    num_bins = int((end-start)/binSize)
    if not num_bins: # A null entry  - can happen with MT and Y
      continue
    
    # Scaled, observed contact matrix, without diagonal
    obs = getContactMatrix(contact_dict, chrA, chrA, limits, limits, binSize).astype(float)
    obs -= diag(diag(obs))
    
    obs_sum = obs.sum()
    if not obs_sum: # Empty matrix
      continue
    
    obs /= obs_sum
    n = len(obs)
    
    # Bin contact data over full extent
    data = contact_dict[chrA][chrA]
    hist, edges = histogram(data[:2].ravel(), bins=num_bins, range=(start, end+binSize))            
    
    if len(hist) < 2:
      continue
    
    # Selection mask for which bin indices are void
    selection = zeros(len(hist), int)
    
    # Set void selection mask based on comparing binned contact data to median
    med = median(hist)

    idx = (hist < (1/clip_factor) * med).nonzero()
    selection[idx] = 1
    
    idx = (hist > (clip_factor) * med).nonzero()
    selection[idx] = 1
    
    # Select close in sequence for extra check - these regions should be strong
    seq_seps = abs(data[0]-data[1])
    
    idx = (seq_seps < close_cis).nonzero()[0]
    close_cis_data = data[:,idx]
    
    # Bin close contacts and set void mask where too sparse
    values = close_cis_data[:2].ravel() # All contact ends, i.e. flatten to diagonal
    c_hist, edges = histogram(values, bins=num_bins, range=(start, end+binSize))
    
    med = median(c_hist)
    idx = (c_hist < (1/clip_factor) * med).nonzero()
    selection[idx] = 1

    # void indices of contact matrix
    idx = selection.nonzero()
    
    # void region ranges
    starts = edges[:-1]
    ends = edges[1:]-1
    
    starts = starts[idx]
    ends = ends[idx]
    
    if len(starts) < 1: # No void in this chromosome
      continue
    
    # Dict of void regions for this chromosome
    regionDict[chrA] = array(zip(starts, ends), int32)
      
  return regionDict
```

```python
def makeAbCompartmentTracks(pop_contacts_path, active_marks_track, binSize=int(2.5e5)):
  """Calculate A/B compartment regions from population Hi-C contacts"""
  
  from numpy import diag, outer, exp, log, cov
    
  chromosomes, chromo_limits, contact_dict = load_contacts(pop_contacts_path)
  marks_regions, marks_values = load_data_track(active_marks_track)
    
  regionDictA = {}
  regionDictB = {}
  
  void_track_regions = calcContactVoidRegions(pop_contacts_path, binSize*4)

  for chromo in chromosomes:

    if chromo not in marks_regions:
      continue
    
    limits = chromo_limits[chromo]
    startPoint, endPoint = limits
    
    # Get observed contacts matrix, remove digonal and scale
    obs = getContactMatrix(contact_dict, chromo, chromo, limits, limits, binSize).astype(float)
    obs -= diag(diag(obs)) # Repeating diag() makes 1D into 2D
    obs /= obs.sum()
    n = len(obs)
    
    # Get total contacts signal and bin counts for each seq separation 
    counts = zeros(n)
    sig = zeros(n)
    
    for i in range(n):
      for j in range(i,n):
        d = j-i
        sig[d] += obs[i,j]
        counts[d] += 1.0
    
    # Get mean contacts signal at each seq sep and normalise to sum 1
    for c, j in enumerate(counts):
      if c:
        sig[j] /= c
    
    sig /= sig.sum()  
    
    # Initial expectation for mean normalised contacts genen the seq. separation
    exp = zeros((n,n), float)
    for i in range(n):
      exp[i,:i+1] = sig[:i+1][::-1]
      exp[i,i:] = sig[:n-i]
    
    # Sum observed rows/cols to get 1D sequence 'sensitivity' values
    vals = obs.sum(axis=0)
    vals /= vals.sum()
    
    # Expectation multiplied prodict of sequence sensitivity values
    exp *= outer(vals, vals)
    exp /= exp.sum()  
   
    # Get seq regions positions for contact map bins
    pos = startPoint + arange(0, n+1) * binSize
    pos = vstack([pos[:-1],  pos[1:]-1]).T.astype(int32)
    
    # Selection mask for non-void regions
    idx_nv = ones(len(pos))
    
    # Zero the void regions of the contact map
    # shifts seq regions about a little to exclude edge cases
    void_regions = array(void_track_regions[chromo], int32)
    
    idx_v = pairRegionsIntersection(pos, void_regions, exclude=False, allow_partial=True)
    idx_nv[idx_v] = 0
    
    # Select only non-void indices
    idx_nv = idx_nv.nonzero()[0]
    obs = obs[idx_nv]
    obs = obs[:,idx_nv]
    exp = exp[idx_nv]
    exp = exp[:,idx_nv]
    
    # Only non-zero values can be used in log ratio
    idx = (exp * obs).nonzero()
    z = ((exp * obs) == 0.0).nonzero()
    
    # Log(obs/exp)
    log_ratio = obs
    log_ratio[idx] /= exp[idx]
    log_ratio[idx] = log(log_ratio[idx])
    log_ratio = np.clip(log_ratio, -4.0, 4.0)
        
    # Covarience matrix
    cov_mat = cov(log_ratio)
    cov_mat[z] = 0.0 
    
    # Cluster covarience into two groups
    centers, clusters, labels = kMeansSpread(cov_mat, 2, verbose=False)
    labels += 1
    
    # Bin active marker data into contact regions
    regions = array(marks_regions[chromo], int32) # start, end
    values = array(marks_values[chromo], float)[:,1] # origValue, normValue
    binned_markers = regionBinValues(regions, values, int32(binSize), startPoint, endPoint)
    
    # Select non-void seq positions and marker data
    pos = pos[idx_nv]
    binned_markers = binned_markers[idx_nv]
    
    # Find group with largest overlap to active marker data
    idx_a = (labels == 1).nonzero()[0]
    idx_b = (labels == 2).nonzero()[0]
    
    f_a = binned_markers[idx_a].sum()/len(idx_a)
    f_b = binned_markers[idx_b].sum()/len(idx_b)
    
    # Swap where needed so "A" compartment overlaps with active markers most
    if f_b > f_a:
      f_a, f_b = f_b, f_a
      idx_a, idx_b = idx_b, idx_a
    
    pos_a = pos[idx_a]
    pos_b = pos[idx_b]
    
    # Fill A/B compartment output dictionaries
    regionDictA[chromo] = pos_a
    regionDictB[chromo] = pos_b
  
  return regionDictA, regionDictB, void_track_regions, chromo_limits

```

```python
def get_data_track_coord_points(coord_path, data_track_path, void_path, model=0,
                                quantile_threshold=None, min_psep=2.0, null_samples=25):
  """Get the particle binned 3D coordinates for a data track in a given structure.
     Save this in a Numpy file for intermediate storage"""
  
  from os.path import join, split, splitext
  from scipy.stats import rankdata
    
  # Load data track
  regions_dict, values_dict = load_data_track(data_track_path)
    
  # Load structure data
  seq_pos_dict, coords_dict = load_structure_coords(coord_path)
  
  chromosomes = set(coords_dict.keys())
    
  # Load data track containing void regions (as calculated by calcContactVoidRegions)
  void_dict = load_data_track(void_path)[0]
  
  # Output file path. Will contain all the 3D coordinate dat for the input data track.
  coord_dir, coord_file = split(coord_path)
  track_dir, track_file = split(data_track_path)
  coord_name = splitext(coord_file)[0]
  track_name = splitext(track_file)[0]  
  track_coord_path = join(coord_dir, '{}.{}.npz'.format(coord_name, track_name))
     
  all_coords = []  # 3D coordinates for binned data tracks across all chromosomes
  all_seq_pos = [] # Positions to calculate sequence separations
  all_values = []  # Binned data track values for above
        
  chromo_ranges = []
  chromo_pos = []
    
  all_coords_null = [[] for i in range(null_samples)]
  all_seq_pos_null = [[] for i in range(null_samples)]
  all_values_null = [[] for i in range(null_samples)]

  max_point = 0   
        
  for chromo in chromosomes:
    void_regions = void_dict[chromo]    # Regions to disregard
    coords = coords_dict[chromo][model] # 3D coordinates for the input structural model
    seq_pos = seq_pos_dict[chromo]      # Seq positions of structure particles
    particle_sep = int32(seq_pos[1]-seq_pos[0]) # Particle bin size
    
    seq_start = seq_pos[0]
    seq_end = seq_pos[-1]
    
    track_values = values_dict[chromo][:,0]
    track_regions = regions_dict[chromo]
      
    # Only accept indices which do not overlap with void regions
    idx = pairRegionsIntersection(track_regions, void_regions, exclude=True)
    track_values  = track_values[idx]
    track_regions = track_regions[idx]
    
    # Make data track regions a point in the middle so it can be allocated to one partile
    middles = track_regions.mean(axis=1)
    track_regions[:,0] = middles
    track_regions[:,1] = middles + 1
    
    # Get seq region to bin over; partile positions are placed in centre of bin
    # Last two particles not used due to disjunction at chromo ends with circular permuted null
    pos_0 = seq_pos[0]  - 0.5 * particle_sep
    
    # Binn the track data to the particle regions
    pos_1 = seq_pos[-1] - (min_psep + 0.5) * particle_sep 
    binned = regionBinValues(track_regions, track_values, np.int32(particle_sep), pos_0, pos_1)
    
    # Select valid particles to use based on binned value
    if (not quantile_threshold) or (track_values.min() == track_values.max()):
      # Consider all non-zero values
      idx_a = binned.nonzero()
 
    else:
      # Only take values above quantile threshold, but weight them equally  
      binned_nz = binned[binned.nonzero()]
      order_a = binned_nz.argsort()
      limit_a = binned_nz[order_a[int(quantile_threshold * len(binned_nz))]]
      idx_a = (binned > limit_a).nonzero()
      track_values = np.ones(track_values.shape, track_values.dtype)
       
    track_coords = coords[idx_a] # 3D coords for the data track in this structure
    track_seq_pos = seq_pos[idx_a].astype(np.int32) # Seq positions for above from the partile bins
    chromo_pos.append(track_seq_pos)
    
    p_max = seq_pos[-1]

    all_seq_pos.append(track_seq_pos.astype(int) + max_point)
    all_coords.append(track_coords)
    all_values.append(binned[idx_a])
    
    # Repeat coordinate extraction for sequence permuted null hypothesis
    for i in range(null_samples):
      j = particle_sep * randint(1, len(coords)-1)
      null_regions = (track_regions + j) % p_max # The permution
      null_values  = np.array(track_values) # Original not binned data
      
      binned = regionBinValues(null_regions, null_values, np.int32(particle_sep), pos_0, pos_1)

      if (not quantile_threshold) or (track_values.min() == track_values.max()):
        idx_n = binned.nonzero()
 
      else: # filtered quantiles
        binned_nz = binned[binned.nonzero()]
        order_n = binned_nz.argsort()
        limit_n = binned_nz[order_n[int(quantile_threshold * len(binned_nz))]]
        idx_n = (binned > limit_n).nonzero()
 
      null_coords = coords[idx_n]
      null_points = seq_pos[idx_n].astype(int)

      all_seq_pos_null[i].append(null_points + max_point)
      all_coords_null[i].append(null_coords) 
      all_values_null[i].append(binned[idx_n]) 
      
    if len(track_seq_pos):
      # For calculation of sequence separations use genome unique
      # with suitably large gaps between each chromosomes
      max_point += track_seq_pos.max() + particle_sep * 5
        
  a = 0
  chromo_ranges = []
  for coords in all_coords:
    b = a + len(coords)
    chromo_ranges.append((a,b))
    a = b

  # Store data as flat arrays that cover all chromosomes
  all_seq_pos = np.concatenate(all_seq_pos, axis=0).astype(np.int64) # Genome unique positions
  all_coords = np.concatenate(all_coords, axis=0)
  all_values = np.concatenate(all_values, axis=0)
  chromo_pos = np.concatenate(chromo_pos, axis=0) # Chromosome sepecific positions
  
  if not quantile_threshold:
    # Quantile normalise binned track values in [0..1]  
    all_values = rankdata(all_values, method='max')
    all_values /= all_values.max()
    
  
  # Flatten null hypothesis data equivalents
  for i in range(null_samples):
    all_seq_pos_null[i] = np.concatenate(all_seq_pos_null[i], axis=0).astype(np.int64)
    all_coords_null[i] = np.concatenate(all_coords_null[i], axis=0)
    all_values_null[i] = np.concatenate(all_values_null[i], axis=0)
    
    if not quantile_threshold:
      all_values_null[i] = rankdata(all_values_null[i], method='max')
      all_values_null[i] /= all_values_null[i].max()
  
  # final null hypothesis data has one row for each sample
  all_seq_pos_null = np.array(all_seq_pos_null)
  all_coords_null = np.array(all_coords_null)
  all_values_null = np.array(all_values_null)
    
  arg_dict = {}
  arg_dict['chromo_ranges'] = chromo_ranges
  arg_dict['chromo_pos'] = chromo_pos
  arg_dict['seq_pos'] = all_seq_pos
  arg_dict['coords'] = all_coords
  arg_dict['values'] = all_values
  arg_dict['seq_pos_null'] = all_seq_pos_null
  arg_dict['coords_null'] = all_coords_null
  arg_dict['values_null'] = all_values_null
    
  np.savez(track_coord_path, **arg_dict)
  
  return track_coord_path
    
```

```python
def calc_spatial_density(coords_points_path_a, coords_points_path_b, min_psep=2, power=2.0):
  """
  Calculate 1/r^2 density for one genome structre of one data track relative to another,
  null adjusted for the second set. Density is calculated per particle point and
  stored as a intermediate .npz file
  """
    
  from os import path
    
  # Names of data tracks and structure
  dir_name_a, file_name_a = path.split(coords_points_path_a)
  dir_name_a, file_name_b = path.split(coords_points_path_b)
  structure_a, track_a = file_name_a.split('.')[:2]
  structure_b, track_b = file_name_b.split('.')[:2]
  
  # Should bee comparing coordinates from the same structure...    
  assert structure_a == structure_b  
    
  # Output Numpy file of spatial density for this pair of data track coordinates
  file_root = path.join(path.dirname(coords_points_path_a), 'density')
  density_file_path = '{}.{}.{}.npz'.format(file_root, track_a, track_b)
  
  #Load Numpy data dictionaries containing coordinate and seq position data fro each data track  
  file_data_a = np.load(coords_points_path_a)
  file_data_b = np.load(coords_points_path_b)
  
  # Start, end and sequence positions for eachchromosomes  
  chromo_ranges = file_data_a['chromo_ranges']
  chromo_pos = file_data_a['chromo_pos']
  particle_sep = int32(chromo_pos[1]-chromo_pos[0])
    
  # Get particle region binned data track positions and values 
  track_coords_a = file_data_a['coords']
  track_values_a = file_data_a['values']
  track_coords_b = file_data_b['coords']
  track_values_b = file_data_b['values']
  
  # Get equivalent permuted null hypothesis values  
  track_coords_n = file_data_b['coords_null']
  track_values_n = file_data_b['values_null']
  
  # Sequence positions required to exclude sequentiall close particles  
  pos_a = file_data_a['seq_pos']
  pos_b = file_data_b['seq_pos']
  pos_n = file_data_b['seq_pos_null']
  min_seq_sep = min_psep * particle_sep * 1.1 
  
  # Density array sizes  
  n = float(len(track_coords_a))
  m = len(track_coords_n)
  
  # Calculate spatial density from inverse square distance summation
  inv_dist_sums = getInvDistSums(track_coords_a, track_coords_b, pos_a, pos_b,
                                 min_seq_sep, track_values_b, power/2)
  density = inv_dist_sums/n

  # Repeat density calculation for permuted null hypothesis
  null_density = []
  for i in range(m):
    inv_dist_sums = getInvDistSums(track_coords_a, track_coords_n[i],
                                   pos_a, pos_n[i], min_seq_sep, track_values_n[i], power/2)
    null_density.append(inv_dist_sums/n)
  
  # Flatten density values over al null samples
  null_density = np.concatenate(null_density)
                                              
  arg_dict = {}
  arg_dict['inv_dist_sums'] = density
  arg_dict['inv_dist_sums_null'] = null_density

  np.savez(density_file_path, **arg_dict)    
  
  return  density_file_path
```

```python
%matplotlib inline
def compare_density_to_random(density_file_path, quantile_weighted=True, n_bins=200, power=2):
  """Plot the spatial density distribution for a pair of data tracks and
     compare to the random expectation derived from permiting the data
     around the seuence for the same structure"""
    
  from matplotlib import pyplot as plt
    
  # Get data track names
  track_a, track_b = density_file_path.split('.')[-3:-1]
  
  # Load density data for a pair of data tracks iin a given structure
  data_dict = np.load(density_file_path)
  density = data_dict['inv_dist_sums']
  null_density = data_dict['inv_dist_sums_null']
  
  # Create plots to display observed density distribution and random expectation
  fig, axarr = plt.subplots(1, 2)
  fig.suptitle('%s to %s' % (track_a, track_b), fontsize=15)
  fig.set_size_inches(12, 5)
  
  if quantile_weighted:
    sde_equation = r'$\frac{i}{N}\sum_{i=1}^{N}{q_i}/r^' + '%d_i$' % power
  else:      
    sde_equation = r'$\frac{1}{N}\sum_{i=1}^{N}1/r^' + '%d_i$' % power
    
  ax = axarr[0]
  ax.set_xlabel(sde_equation)
  ax.set_ylabel('Probability mass')
  
  # start binning at minimum value
  bin_min = min(density.min(), null_density.min())
  
  # There can be outliers stop binning at largest median + 3 sigma
  # This carrys the bulk of the information and ensures a good bin
  # density over the most important region
  bin_max = max(np.median(density) + 3 * density.std(),
                np.median(null_density) + 3 * null_density.std())
  
  ax.set_xlim((bin_min, bin_max))
  
  # create histograms of density distribution and its null equivalent
  hist_obs, edges_obs = np.histogram(density, bins=n_bins, range=(bin_min, bin_max), normed=True)
  hist_exp, edges_exp = np.histogram(null_density, bins=n_bins, range=(bin_min, bin_max), normed=True)
  
  # Sum to 1.0 for entropy calculation (pseudo-probabilities) 
  hist_obs /= hist_obs.sum()
  hist_exp /= hist_exp.sum()
  
  # Plot line graphs  
  ax.plot(edges_obs[:-1], hist_obs, label='Obs.', color='#0080FF')
  ax.plot(edges_exp[:-1], hist_exp, label='Exp.', color='#808080')
  ax.legend(fontsize='small')

  ax = axarr[1]
  ax.set_xlabel(sde_equation)
  ax.set_ylabel(r'Enrichment: $P_{obs}.\log_2(\frac{P_{obs}}{P_{exp}})$')
  ax.set_xlim((bin_min, bin_max))
  
  # Calculate Kullback_liebler divergence for nonzero histogram values
  # as a measure of information difference from the random hypothesis
  nz = (hist_obs * hist_exp).nonzero()
  x_vals = edges_exp[:-1]
  h = np.zeros(len(hist_obs), float)
  h[nz] = hist_obs[nz] * np.log2(hist_obs[nz]/hist_exp[nz])   

  ax.plot(x_vals, h, color='#FF8000')
  ax.hlines(0, bin_min, bin_max, color='#808080', alpha=0.5)
  ax.set_title('Information gain over random (H=%.3f)' % h.sum(), fontsize=12)
  
  relative_entropy = h.sum()

  print("Ralative entropy %s - %s: %f" % (track_a, track_b, relative_entropy))
  
  plt.show()
```
